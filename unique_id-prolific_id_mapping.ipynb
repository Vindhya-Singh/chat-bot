```python
import json
import pandas as pd
results_folder = f'/projects/ReflectiveLearningChatbots/results_run1'
chatbot_experiment = '3_jap'
with open(results_folder + f'/data_{chatbot_experiment}.json', 'r') as f:
    all_data = json.load(f)
participants = all_data['participants']
```

```python
prolific_ids = []
unique_ids = []
for part in range(len(participants)):
    prolific_id = participants[part]['parameters']['company']
    unique_id = participants[part]['uniqueId']
    prolific_ids.append(prolific_id)
    unique_ids.append(unique_id)
    
print(len(prolific_ids))
print(len(unique_ids))
```

```python
prolific_ids
```

```python
df_new = pd.DataFrame({'uniqueId':unique_ids,"Prolific_Ids":prolific_ids})
df_new.head()
```

```python
df_new.to_csv(f'uniqueIds_to_ProlificIds_{chatbot_experiment}.csv', index=False)
```
